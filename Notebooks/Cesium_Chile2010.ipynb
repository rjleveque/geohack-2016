# Cesium_Chile2010

Illustrates how to plot kml files on top of maps or imagery in Cesium.

This example uses output files generated by a [GeoClaw simulation of the 2010 tsunami originating near Maule, Chile.](http://www.clawpack.org/_static/geoclaw/examples/tsunami/chile2010/README.html)  This is illustrated in [this animation](http://www.clawpack.org/_static/geoclaw/examples/tsunami/chile2010/_plots/movieframe_allframesfig0.html). 

Here we plot a few frames from the animation using Cesium.

```python
import cesiumpy  # install from https://pypi.python.org/pypi/cesiumpy
```

```python
v = cesiumpy.Viewer()

frameno = 6
kml_file = '../DataFiles/Chile_2010_kml/fig1/frame%sfig1/doc.kml' % str(frameno).zfill(4)
print 'Will plot %s' % kml_file
v.dataSources.add(cesiumpy.KmlDataSource(kml_file))

viewpoint = (-70, -30, 1e7)  # inital view (longitude, latitude, elevation)
v.camera.flyTo(viewpoint)
v
```

In Google Earth it is possible to open either `kmz_file` or `kml_file2` below to open all frames and a slider bar to show the animation.  This does not appear to work in Cesium.

```python
v = cesiumpy.Viewer()

kmz_file = '../DataFiles/Chile_2010_kml/Chile_2010.kmz' 
kml_file2 = '../DataFiles/Chile_2010_kml/fig1/doc.kml' 

v.dataSources.add(cesiumpy.KmlDataSource(kml_file2))

viewpoint = (-70, -30, 1e7)  # inital view (longitude, latitude, elevation)
v.camera.flyTo(viewpoint)
v
```
